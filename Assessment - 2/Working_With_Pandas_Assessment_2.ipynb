```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("/content/Retail Data.csv")
data
```

```python
data.columns
```

```python
# 1. View the structure of the dataset (columns, types, missing values).

data.info()
data.isnull().sum()
```

```python
# 2. What is the shape (rows, columns) of the dataset?

data.shape
```

```python
# 3. Are there any duplicate records?

data.duplicated().sum()
```

```python
# 4. Are there any missing or corrupted entries in Ship Date, Order Date, or numeric columns?

display(data.isnull().sum())
```

```python
# 5. Convert Order Date and Ship Date to datetime.

data['Order Date'] = pd.to_datetime(data['Order Date'], format='%d-%m-%Y')
data['Ship Date'] = pd.to_datetime(data['Ship Date'], format='%d-%m-%Y')
data.info()
```

```python
# 6. Check for future or inconsistent shipping dates.

data[data['Ship Date'] < data['Order Date']]
```

```python
# 7. Convert price columns to numeric (remove $ and commas).

data['Cost Price'] = data['Cost Price'].astype(str).str.replace('$', '', regex=False).str.replace(',', '', regex=False).astype(float)
data['Retail Price'] = data['Retail Price'].astype(str).str.replace('$', '', regex=False).str.replace(',', '', regex=False).astype(float)
data
```

```python
# 8. What are the unique values in Customer Type and Order Priority?

print(data['Customer Type'].unique())
print()
print(data['Order Priority'].unique())
```

```python
# 9. What are the most common shipping modes?

data['Ship Mode'].value_counts()
```

```python
# 10. Which cities have the highest number of orders?

data['City'].value_counts().head()
```

```python
# 11. Whatâ€™s the range of order quantities and prices?

print("Order Quantity:", (data['Order Quantity'].min(), data['Order Quantity'].max()))
print("Retail Price :", (data['Retail Price'].min(), data['Retail Price'].max()))
print("Cost Price :", (data['Cost Price'].min(), data['Cost Price'].max()))
```

```python
# 12. Create a new column for shipping duration.

data['Shipping Duration'] = data['Ship Date'] - data['Order Date']
data['Shipping Duration']
```

```python
# 13. Are there any orders with zero or negative total or quantity?

data['Total'] = data['Total'].astype(str).str.replace('$', '', regex=False).str.replace(',', '', regex=False).astype(float)

print(data[data['Total'] <= 0])
print()
print(data[data['Order Quantity'] <= 0])
```

```python
# 14. Are all discount percentages matching discount dollar amounts?

data[data['Discount %'] != data['Discount $']]
```

```python
# 15. Check for mismatches in total calculationâ€™.

data['total_calculation'] = data['Order Quantity'] * data['Retail Price']
data[data['Total'] != data['total_calculation']]
```

```python
# 16. Identify top 5 products by order quantity.

data.groupby('Product Name')['Order Quantity'].sum().head()
```

```python
# 17. Which Account Manager handled the most revenue?

data.groupby('Account Manager')['Total'].sum().sort_values(ascending=False).head(1)
```

```python
# 18. What is the average shipping cost by mode?

if data['Shipping Cost'].dtype == 'object':
    data['Shipping Cost'] = data['Shipping Cost'].astype(str).str.replace('$', '', regex=False).str.replace(',', '', regex=False).astype(float)

data.groupby('Ship Mode')['Shipping Cost'].mean()
```

```python
# 19. Find the most profitable product.

data.groupby('Product Name')['Total'].sum().sort_values(ascending=False).head(1)
```

```python
# 21. Which customer type generates more revenue?

data.groupby('Customer Type')['Total'].sum().sort_values(ascending=False).head()
```

```python
# 22. How does order priority affect revenue?

data.groupby('Order Priority')['Total'].sum().sort_values(ascending=False)
```

```python
# 23. What is the average profit margin by product category?

if data['Profit Margin'].dtype == 'object':
    data['Profit Margin'] = data['Profit Margin'].astype(str).str.replace('$', '', regex=False).str.replace(',', '', regex=False).astype(float)

av = data.groupby('Product Category')['Profit Margin'].mean()
display(av)
```
