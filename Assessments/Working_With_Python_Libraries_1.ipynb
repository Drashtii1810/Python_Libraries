```python
#What are the top 5 states with the most failed banks?

bank['ST'].value_counts().head(5)
```

```python
#What are the top 5 acquiring institutions?

bank['Acquiring Institution'].value_counts().head(5)
```

```python
#How many banks has the State Bank of Texas acquired?

(bank['Acquiring Institution'] == 'State Bank of Texas').sum()
```

```python
#How many of them were actually in Texas?

bank[(bank['Acquiring Institution'] == 'State Bank of Texas') & (bank['ST'] == 'TX')].shape[0]
```

```python
#What is the most common city in California for a bank to fail in?

bank[bank['ST'] == 'CA']['City'].value_counts().idxmax()
```
